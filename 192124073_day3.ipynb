#**17.Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school.**

```python
import pandas as pd
file_path = 'school_data.csv'
df = pd.read_csv(file_path)
result = df.groupby('school')['age'].agg(['mean', 'min', 'max']).reset_index()
print(result)
```

#**18.Write a Pandas program to split the following given dataframe into groups based on school code and class.**

```python
import pandas as pd
df = pd.read_csv('school_data.csv')
grouped = df.groupby(['school', 'class'])
for group_name, group_data in grouped:
    print(f"Group: {group_name}")
    print(group_data)
    print("\n")
```

#**20.Write a Pandas program to find the index of a given substring of a DataFrame column**

```python
import pandas as pd
data = {'Column1': ['Hello, World!', 'Python is awesome', 'Data Analysis']}
df = pd.DataFrame(data)
substring = 'World'
df['Index'] = df['Column1'].str.find(substring)
print(df)
```

#**21.Write a Pandas program to swap the cases of a specified character column in a given DataFrame.**

```python
import pandas as pd
data = {'Name': ['John', 'Alice', 'Bob', 'Eve'],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Miami']}
df = pd.DataFrame(data)
column_to_swap_case = 'City'
df[column_to_swap_case] = df[column_to_swap_case].str.swapcase()
print(df)
```

#**24.Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.**
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('financial_data.xlsx')
dates = df['Date']
high_prices = df['High']
low_prices = df['Low']
close_prices = df['Close']
plt.figure(figsize=(10, 6))
plt.plot(dates, high_prices, label='High', marker='o')
plt.plot(dates, low_prices, label='Low', marker='o')
plt.plot(dates, close_prices, label='Close', marker='o')
plt.title('Alphabet Inc. Stock Prices (Oct 3, 2016 - Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%m-%d-%y'))
plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.DayLocator(interval=1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#**25.Write a Python program to plot two or more lines with legends, different widths and colors.**

#**26.Write a Python program to create multiple plots.**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Function')
axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Function')
axes[1, 0].plot(x, y3)
axes[1, 0].set_title('2 * Sine Function')
axes[1, 1].plot(x, y4)
axes[1, 1].set_title('2 * Cosine Function')
plt.tight_layout()
plt.show()
```
